```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import gamma
```

```python
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

n = pd.Series(range(1,30))

y1 = 0*n + 1
y2 = np.log(n)
y3 = n
y4 = n * np.log(n)
y5 = n**2
y6 = 2**n
y7 = n.apply(np.math.factorial)
y8 = n**n


plt.plot(n,y1, label = 'O(1)')
plt.plot(n,y2, label = 'Olog(n)')
plt.plot(n,y3, label = 'On(n log n)')
plt.plot(n,y4, label = 'O(n)')
plt.plot(n,y5, label = 'O(n^2)')
plt.plot(n,y6, label = 'O(2^n)')
plt.plot(n,y7, label = 'O(n!)')
plt.plot(n,y8, label = 'O(n^n)')

plt.xlim(0,20)
plt.ylim(0,20)

plt.legend()
plt.show()
```
