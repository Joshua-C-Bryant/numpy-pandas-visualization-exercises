```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Use matplotlib to plot the following equation: y = x^2 - x + 2

```python
x = pd.Series(range(-51,51))

def x_squared_minus_x_plus_2(x):
    return x**2 - x + 2

y = x.apply(x_squared_minus_x_plus_2)
```

```python
plt.plot(x,y)
plt.annotate('Intersection', xy=(0,0), xytext=(0,500),arrowprops={'facecolor':'green'})
plt.title('$x^2$ - x + 2')
plt.show()
```

### Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

#### y = √x

```python
import math

x = pd.Series(range(0,51))
def square_root(x):
    if x in range(0,51):
        return math.sqrt(x)
y = x.apply(square_root)
```

```python
plt.plot(x,y)
plt.title('Square Root of X')
plt.show()
```

#### x^3

```python
x = pd.Series(range(-50,51))
def x_cubed(x):
    return x**3
y = x.apply(x_cubed)
```

```python
plt.plot(x,y)
plt.title('$x^3$')
plt.show()
```

#### 2^x

```python
x = pd.Series(range(0,51))
def two_raised_to_the(x):
    if x in range(0,51):
        return 2**x
y = x.apply(two_raised_to_the)
```

```python
plt.plot(x,y)
plt.title('$2^x$')
plt.show()
```

#### 1/(x+1)

```python
x = pd.Series(range(0,51))
def one_over_x_plus_one(x):
    if x in range(0,51):
        return 1/(x+1)
y = x.apply(one_over_x_plus_one)
```

```python
plt.plot(x,y)
plt.title('1/(x+1)')
plt.show()
```

### Combine the figures you created in the last step into one large figure with 4 subplots.



```python
x = pd.Series(range(0,50))

y1 = x.apply(square_root)
y2 = x.apply(x_cubed)
y3 = x.apply(two_raised_to_the)
y4 = x.apply(one_over_x_plus_one)

plt.subplot(221)
plt.plot(x, y1)
plt.title('Square Root of X')

plt.subplot(222)
plt.plot(x, y2)
plt.title('$x^3$')

plt.subplot(223)
plt.plot(x, y3)
plt.title('$2^x$')

plt.subplot(224)
plt.plot(x, y4)
plt.title('1/(x+1)')

plt.tight_layout
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.35)
plt.show()
```

### Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = pd.Series(range(0,50))

y1 = x.apply(square_root)
y2 = x.apply(x_cubed)
y3 = x.apply(two_raised_to_the)
y4 = x.apply(one_over_x_plus_one)

plt.ylim(top=50)
plt.xlim(0,50)

plt.plot(x, y1, label = 'Square Root of X')
plt.plot(x, y2, label = '$x^3$')
plt.plot(x, y3, label = '$2^x$')
plt.plot(x, y4, label = '1/(x+1)')

plt.legend()
plt.show()
```
