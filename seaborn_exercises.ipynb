```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.dtypes
```

```python
iris.head()
```

```python
plt.title('Distribution of Petal Lengths')
sns.histplot(x='petal_length',data=iris)
plt.xlabel('Petal Length')
plt.show()
```

```python
sns.relplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title('Petal Length vs Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
sns.relplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
# Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
from env import host, user, password
url = f'mysql+pymysql://{user}:{password}@{host}/chipotle'
```

```python
pd.read_sql('SELECT * FROM orders', url)
```

```python
orders_sql = '''
SELECT *
FROM orders
'''
orders = pd.read_sql(orders_sql, url)
orders.head()
```

```python
orders['item_price'] = orders.item_price.astype(str)
orders['item_price'] = orders.item_price.str.replace('$', '', regex=False).astype('float')
orders['item_price']
```

```python
top_sellers = orders.groupby(['item_name']).quantity.sum().sort_values(ascending=False).nlargest(4)
```

```python
revenue = orders[orders.item_name.isin(top_sellers.index)].groupby('item_name').item_price.sum()
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
sns.barplot(y='item_name',x='item_price',data=revenue)
plt.title('Most Popular Items and Revenue')
plt.ylabel('Item')
plt.xlabel('Revenue')
plt.show()
```

```python
# Load the sleepstudy data and read it's documentation. 
# Use seaborn to create a line chart of all 
# the individual subject's reaction times and 
# a more prominant line showing the average change in reaction time.
sleepstudy = data('sleepstudy')
sleepstudy
sleepstudy['Subject'] = 'Subject_' + sleepstudy.Subject.astype(str)
plt.figure(figsize=(15,10))
sns.lineplot(x='Days',y='Reaction',hue='Subject',data=sleepstudy)
sns.lineplot(x='Days',y='Reaction',color ='black',estimator='mean',data=sleepstudy)
plt.show()
```
